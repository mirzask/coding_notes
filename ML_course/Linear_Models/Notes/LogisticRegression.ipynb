# Logistic Regression/Classification

Create X and y

```python
X = data.iloc[:,:2].values # values from columns 0 and 1
y = data.iloc[:,2].values # values from the 'released' column
```

Create a function to display the separating curve classifier:

```python
def plot_boundary(clf, X, y, grid_step=.01, poly_featurizer=None):
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, grid_step),
                         np.arange(y_min, y_max, grid_step))


    # to every point from [x_min, m_max]x[y_min, y_max]
    # we put in correspondence its own color
    Z = clf.predict(poly_featurizer.transform(np.c_[xx.ravel(), yy.ravel()]))
    Z = Z.reshape(xx.shape)
    plt.contour(xx, yy, Z, cmap=plt.cm.Paired)
```

## Polynomial pre-processing

Pre-processing with `PolynomialFeatures`. This will give a matrix of all the possible polynomial combinations <= specified polynomial

```python
poly = PolynomialFeatures(degree=7)
X_poly = poly.fit_transform(X)

X_poly.shape
```

## Build Logistic Regression model - $C = 10^-2$

Train a **logistic regression** model using a regularization parameter $C = 10^-2$

```python
C = 1e-2
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y)

plot_boundary(logit, X, y, grid_step=.01, poly_featurizer=poly)

plt.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', label='Released')
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='orange', label='Faulty')
plt.xlabel("Test 1")
plt.ylabel("Test 2")
plt.title('2 tests of microchips. Logit with C=%s' % C)
plt.legend();

print(f'Accuracy on training set: {round(logit.score(X_poly, y), 3)}')
```

## Build Logistic Regression model - $C = 1$

Train a **logistic regression** model using a regularization parameter $C = 1$

```python
C = 1
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y)

plot_boundary(logit, X, y, grid_step=.01, poly_featurizer=poly)

plt.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', label='Released')
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='orange', label='Faulty')
plt.xlabel("Test 1")
plt.ylabel("Test 2")
plt.title('2 tests of microchips. Logit with C=%s' % C)
plt.legend();

print(f'Accuracy on training set: {round(logit.score(X_poly, y), 3)}')
```

## Build Logistic Regression model - $C = 10,000$

Train a **logistic regression** model using a regularization parameter $C = 10^4$

```python
C = 1e4
logit = LogisticRegression(C=C, random_state=17)
logit.fit(X_poly, y)

plot_boundary(logit, X, y, grid_step=.01, poly_featurizer=poly)

plt.scatter(X[y == 1, 0], X[y == 1, 1], c='blue', label='Released')
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='orange', label='Faulty')
plt.xlabel("Test 1")
plt.ylabel("Test 2")
plt.title('2 tests of microchips. Logit with C=%s' % C)
plt.legend();

print(f'Accuracy on training set: {round(logit.score(X_poly, y), 3)}')
```

## Hyperparameter tuning of $C$

Use `LogisticRegressionCV` to perform grid search of parameters $\rightarrow$ cross-validation for Logistic Regression models.

> For other models, you will need to use `GridSearchCV`, `RandomizedSearchCV`, etc.

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)

c_values = np.logspace(-2, 3, 500)

logit_searcher = LogisticRegressionCV(Cs=c_values, cv=skf, verbose=1, n_jobs=-1)
logit_searcher.fit(X_poly, y)
```

```python
logit_searcher.C_
```

How does the quality of the model vary by changing the $C$ hyperparameter?

Answet this be creating a validation curve

```python
plt.plot(c_values, np.mean(logit_searcher.scores_[1], axis=0))
plt.xlabel('C')
plt.ylabel('Mean CV-accuracy');
```
